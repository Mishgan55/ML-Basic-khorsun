## Задание 1: Конвертер регистров

Написать функцию, которая будет переводить snake_case в PascalCase и наоборот. 

Функция должна сама определять - какой формат ей передали. Можно добавить ключевой аргумент, который будет принудительно возвращать один из форматов.

<br>

**Примеры:**
* `otus_course     -> OtusCourse`
* `PythonIsTheBest -> python_is_the_best`



```python
# ВАШ КОД ЗДЕСЬ
request = input()


def convert_to_pascal_or_snake_case(text: str) -> str:
  if "_" in text:
    return "".join(word.capitalize() for word in text.split("_"))
  else:
    result = text[0].lower()
    for index, char in enumerate(text[1:]):
      if char.isupper():
        result = result + "_" + char.lower()
      else:
        result += char
    return result


convert_to_pascal_or_snake_case(request)
```

## Задание 2: Проверка валидности даты

Написать функцию проверяющую валидность введенной даты.

<br>

**Примеры:**
* `29.02.2000 -> True`
* `29.02.2001 -> False`
* `31.04.1962 -> False`



```python
# ВАШ КОД ЗДЕСЬ

request = input()


def is_date_valid(date: str) -> bool:
  arrays = date.split(".")
  day, month, year = arrays[0], arrays[1], arrays[2]
  return is_month_valid(month) and is_day_valid(day, month, is_ves_year(year))


def is_ves_year(year: str) -> bool:
  return int(year) % 400 == 0


def is_month_valid(month: str):
  return int(month) <= 12


def is_day_valid(day: str, month: str, is_ves: bool) -> bool:
  max_days = {"01": 31,
              "02": 29 if is_ves else 28,
              "03": 31,
              "04": 30,
              "05": 31,
              "06": 30,
              "07": 31,
              "08": 31,
              "09": 30,
              "10": 31,
              "11": 30,
              "12": 31}

  return int(day) <= max_days.get(month)


print(is_date_valid(request))
```

## Задание 3: Проверка на простое число

Функция проверки на простое число. Простые числа – это такие числа, которые делятся на себя и на единицу.

<br>

**Примеры:**
* `17 -> True`
* `20 -> False`
* `23 -> True`

```python
# ВАШ КОД ЗДЕСЬ
import math

request = int(input())


def is_number_simple(number: int) -> bool:
  if number <= 1:
    return False
  if number == 2:
    return True
  if number % 2 == 0:
    return False
  else:
    sqrt_number = int(math.sqrt(number))
    is_simple = True
    for i in range(2, sqrt_number):
      if number % i == 0:
        is_simple = False
    return is_simple


print(is_number_simple(request))
```

## Задание 4: Учет пользователей

Пользователь в бесконечном цикле вводит данные пользователей: имя, затем фамилию, возраст и ID. Ввод продолжается до тех пор, пока не будет введено пустое поле. 

Пользователи заносятся в словарь, где ключ это ID пользователя, а остальные данные записываются в виде кортежа. 

**Программа должна проверять:**
* имя и фамилия состоят только из символов и начинаются с большой буквы - если не с большой, то заменяет букву на большую;
* возраст должен быть числом от 18 до 60;
* ID - целое число, дополненное до 8 знаков незначащими нулями, ID должен быть уникальным.

**Дополнительно:** написать функцию, которая будет выводить полученный словарь в виде таблицы.

```python
def format_name(name: str) -> str:
    if not name.isalpha():
        raise ValueError("Имя и фамилия должны содержать только буквы")
    return name.capitalize()

def print_users_table(users: dict):
    for user_id, data in users.items():
        name, surname, age = data
        print(f"{user_id:<10} {name:<15} {surname:<15} {age:<6}")


def main():
    users = {}

    while True:
        name = input("Введите имя (пусто для выхода): ").strip()
        if not name:
            break

        surname = input("Введите фамилию: ").strip()
        if not surname:
            break

        age_input = input("Введите возраст: ").strip()
        if not age_input:
            break

        id_input = input("Введите ID: ").strip()
        if not id_input:
            break

        try:
            name = format_name(name)
            surname = format_name(surname)
        except ValueError as e:
            print(e)
            continue

        if not age_input.isdigit():
            print("Возраст должен быть числом")
            continue
        age = int(age_input)
        if age < 18 or age > 60:
            print("Возраст должен быть от 18 до 60")
            continue

        if not id_input.isdigit():
            print("ID должен быть числом")
            continue
        user_id = id_input.zfill(8)
        if user_id in users:
            print("Этот ID уже существует, введите другой")
            continue

        users[user_id] = (name, surname, age)
    print_users_table(users)

main()
```
