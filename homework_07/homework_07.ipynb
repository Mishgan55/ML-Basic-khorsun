### Создайте два двумерных массива размерами 6x4 и 4x3 и произведите их матричное умножение.

```python
first_arr = np.random.random((6, 4))
# print(first_arr)
second_arr = np.random.random((4, 3))
# print(second_arr)

result = first_arr @ second_arr
print(result)
```

### Создайте случайный двумерный массив 7x7, найти у него среднее и стандартное оклонение. Нормализуйте этот массив.

```python

arr = np.random.randint(0, 100, size=(7, 7))

mean = np.mean(arr)
std = np.std(arr)

arr_normalized = (arr - mean) / std

print("Массив:\n", arr)
print("Среднее:", mean)
print("Стандартное отклонение:", std)
print("Нормализованный массив:\n", arr_normalized)

```

```python
result = (tips['smoker'] == 'Yes').sum()
print(result)
```

### Узнайте какой средний 'total_bill' в зависимости от 'day'

```python
tips.groupby('day')['total_bill'].mean()
```

### Отберите строчки с 'total_bill' больше медианы и узнайте какой средний 'tip' в зависимости от 'sex'

```python
median_total_bill = tips['total_bill'].median()
# print(median_total_bill)

total = tips[tips['total_bill'] > median_total_bill]
# print(total)

result = total.groupby('sex')['tip'].mean()

print(result)
```
